```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns


# Загрузка данных
data = pd.read_csv('restaurant_data.csv')

print(data.head())
```

```python
numerical_columns = [
    'Rating', 'Seating Capacity', 'Average Meal Price', 'Marketing Budget',
    'Social Media Followers', 'Chef Experience Years', 'Number of Reviews',
    'Avg Review Length', 'Ambience Score', 'Service Quality Score',
    'Weekend Reservations', 'Weekday Reservations', 'Revenue'
]

# Вычисление корреляционной матрицы
corr_matrix = data[numerical_columns].corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


# Функция для построения регрессии и визуализации
def plot_regression(x_col, y_col, data):
    X = data[[x_col]]
    y = data[y_col]
    
    reg = LinearRegression().fit(X, y)
    y_pred = reg.predict(X)
    
    r2 = r2_score(y, y_pred)
    print(f'R² для {x_col} и {y_col}: {r2}')
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_col, y=y_col, data=data)
    plt.plot(data[x_col], y_pred, color='red', linewidth=2)
    plt.title(f'Регрессия: {x_col} и {y_col}')
    plt.show()

# Построение регрессий для указанных пар показателей
pairs = [
    ('Seating Capacity', 'Revenue'),
    
]

for x_col, y_col in pairs:
    plot_regression(x_col, y_col, data)
```
